```
Raw_Data = """Gender,Age,Income,Education,Employement,Subscription
Male,17,0,HS,Student,Yes
Male,26,32,PG,Manager,Yes
Female,29,37,PG,Nurse,Yes
Female,22,12,UG,Student,No
Male,26,32,Diploma,Farmer,Yes
Female,35,29,Diploma,Nurse,Yes
Female,41,47,PG,Farmer,No
Male,20,0,HS,Student,No
Male,39,42,PG,Manager,Yes
Female,50,38,UG,Nurse,Yes"""
f = open('Q1.csv','w')
f.write(Raw_Data)
f.close()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_text,plot_tree
```

```
Dataset = pd.read_csv('Q1.csv')
Dataset
```

```
Data = np.array(Dataset)
Features = Data[:,:-1]
Target = Data[:,-1]
print(Features)
print(Target)
```

```
Gender,Education,Employement = LabelEncoder(),LabelEncoder(),LabelEncoder()
Features[:,0] = Gender.fit_transform(Features[:,0])
Features[:,3] = Education.fit_transform(Features[:,3])
Features[:,4] = Employement.fit_transform(Features[:,4])
print(Features)
```

```
Subscription = LabelEncoder()
Target = Subscription.fit_transform(Target)
print(Target)
print(Subscription.inverse_transform(Target.astype('int')))
```

```
model = DecisionTreeClassifier()
model.fit(Features,Target)
model.score(Features,Target)
```

```
print('Text Representation Of Tree : \n')
print(export_text(model,feature_names=list(Dataset)[:-1]))
```

```
plt.figure(figsize=(9,9))
plot_tree(model,feature_names=list(Dataset),filled=True,rounded=True,node_ids=True)
plt.show()
```

```
print('Targetted Values : ')
print(Target)
print('Model Predicted Values : ')
print(model.predict(Features))
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
X = np.array([49,63,58,60,58,61,63,52])
Y = np.array([15,23,18,23,24,22,22,19])
Target = np.array([65])
```

```
plt.figure()
sns.scatterplot(x=X,y=Y)
plt.show()
```

```
model = LinearRegression()
model.fit(X.reshape(-1,1),Y.reshape(-1,1))
```

```
plt.figure()
plt.scatter(x=X,y=Y)
Line = np.array(model.predict(X.reshape(-1,1)))
plt.plot(X.reshape(-1,1),Line.reshape(-1,1),c='r')
plt.scatter(Target,model.predict(Target.reshape(-1,1)))
plt.show()
print('Target Value : ',Target)
print('Model Predicted Value : ',model.predict(Target.reshape(-1,1)))
```

```
def Associations(LIST):
  for a in LIST:
      #print(a)
      items = [x for x in a.items]
      for x in range(len(items)-1):
        print(items[x],end=' <-> ')
      print(items[len(items)-1])
      print('Support : ',a[1])
      for x in a[2]:
        for a1 in x[0]:
          for a2 in x[1]:
            print(a1,'->',a2,end=' -- ')
        print('Confidence : ',x[2],end=' -- ')
        print('Lift : ',x[3])
        print('--------------------------')
```

```
model = apyori.apriori(ItemList)
Associations(model)
```

```
model = apyori.apriori(ItemList,min_confidence=0.6,min_support=0.5)
Associations(model)
```
