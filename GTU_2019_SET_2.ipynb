```
Raw_Q1 = """Day Outlook Temp Humidity Windy Play
1 Sunny 85 85 Weak No
2 Sunny 80 90 Strong No
3 Overcast 83 86 Weak Yes
4 Rainy 70 96 Weak Yes
5 Rainy 68 80 Weak Yes
6 Rainy 65 70 Strong No
7 Overcast 64 65 Strong Yes
8 Sunny 72 95 Weak No
9 Sunny 69 70 Weak Yes
10 Rainy 75 80 Weak Yes
11 Sunny 75 70 Strong Yes
12 Overcast 72 90 Strong Yes
13 Overcast 81 75 Weak Yes
14 Rainy 71 91 Strong No"""

f = open('Q1.tsv','w')
f.write(Raw_Q1)
f.close()
```

```
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.preprocessing import LabelEncoder
```

```
Dataset = pd.read_csv('Q1.tsv',sep=' ')
Dataset
```

```
Data = np.array(Dataset)
Data
Feature = Data[:,1:-1]
Target = Data[:,-1]
```

```
Outlook,Windy,Play = LabelEncoder(),LabelEncoder(),LabelEncoder()
Feature[:,0] = Outlook.fit_transform(Feature[:,0])
Feature[:,3] = Windy.fit_transform(Feature[:,3])
Target = Play.fit_transform(Target)
```

```
model = DecisionTreeClassifier()
model.fit(Feature,Target)
model.score(Feature,Target)
```

```
plt.figure(figsize=(9,9))
plot_tree(model,feature_names=['Outlook','Temp','Humidity','Windy','Play'],filled=True)
plt.show()
```

```
Q3_Data = """X Y Class
4.4 3.2 Iris-setosa
5 3.5 Iris-setosa
5.1 3.8 Iris-setosa
4.8 3 Iris-setosa
5.1 3.8 Iris-setosa
4.6 3.2 Iris-setosa
5.3 3.7 Iris-setosa
5 3.3 Iris-setosa
7 3.2 Iris-versicolor
6.4 3.2 Iris-versicolor
6.9 3.1 Iris-versicolor
5.5 2.3 Iris-versicolor
6.5 2.8 Iris-versicolor
5.7 2.8 Iris-versicolor
6.3 3.3 Iris-versicolor
4.9 2.4 Iris-versicolor
6.6 2.9 Iris-versicolor"""
f = open('Q3.tsv','w')
f.write(Q3_Data)
f.close()
```

```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
Dataset = pd.read_csv('Q3.tsv',sep=' ')
Dataset
```

```
Data = np.array(Dataset)
Data
```

```
model = KMeans(n_clusters=2,n_init=10,max_iter=300,tol=1e-04,random_state=1,init='random',n_jobs=4,precompute_distances=True)
model.fit(Data[:,:2])
```

```
Centroids = np.array(model.cluster_centers_)
plt.figure(figsize=(7,7))
plt.scatter(Data[:,0],Data[:,1],c=model.labels_,cmap='rainbow')
plt.scatter(Centroids[:,0],Centroids[:,1],marker='X',linewidths=2)
plt.show()
```
