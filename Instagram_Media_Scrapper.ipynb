```
PROFILE = 'junaid_movar' #input('Enter Instagram Username Without @ : ')
import os
import matplotlib.pyplot as plt
%matplotlib inline
os.system('pip install instaloader')
if os.system('instaloader '+PROFILE+' -q') == 2:
  print('Unable To Find Profile')
```

```
Data = os.listdir(PROFILE)
for _ in range(len(Data)):
  Data[_] = PROFILE+'/'+Data[_]
```

```
Images = []
for _ in range(len(Data)):
  if Data[_].endswith('jpg') and not Data[_].endswith('profile_pic.jpg'):
    Images.append(Data[_])
```

```
DP = ''
for _ in Data:
  if _.endswith('profile_pic.jpg'):
    DP = _
plt.figure(figsize=(6,6))
plt.imshow(plt.imread(DP))
plt.title('PROFILE PICTURE')
plt.xticks([])
plt.yticks([])
plt.show()
```

```
count = 0
while count < len(Images):
  Fig, Sub = plt.subplots(1,3,figsize=(10,10))
  for _ in range(3):
    Sub[_].imshow(plt.imread(Images[count]))
    Sub[_].set_xticks([])
    Sub[_].set_yticks([])
    count+=1
  Fig.show()
```
